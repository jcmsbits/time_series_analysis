The QQ Plot

Quantile - Quantile Plot (QQ Plot)
Used to determine whether a data set is distributed a certain way
Usually showcases how the data fits a Normal Distribution

```python
import scipy.stats
import pylab
```

Takes all the values a variable can take, and arranges them in  accending order

```python
df_comp.max()
```

```python
# El eje y representa el precio
# El eje x representa "How many standard deviations away from the mean these values are"
scipy.stats.probplot(df_comp.spx, plot = pylab)
pylab.show()
```

```python
scipy.stats.probplot(df_comp.dax, plot = pylab)
pylab.show()
```

```python
scipy.stats.probplot(df_comp.ftse, plot = pylab)
pylab.show()
```

```python
scipy.stats.probplot(df_comp.nikkei, plot = pylab)
pylab.show()
```
